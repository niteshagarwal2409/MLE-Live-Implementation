Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Creating Randomised Dataset

```python
X = 3*np.random.rand(100,1)
```

```python
Y= 9+2*X + np.random.rand(100,1)
```

```python
plt.scatter(X,Y)
plt.show()
```

Calculating Mean of X and Y

```python
mean_X = X.mean()
mean_Y = Y.mean()
```

Calculating Beta estimates for Linear Regression

```python
B1 = ((X-mean_X)*(Y-mean_Y)).sum()/((X-mean_X)**2).sum()
```

B0= mean_Y-B1*mean_X

```python
print(B0,B1)
```

```python
plt.scatter(X,Y)
plt.plot(X,B0+B1*X)
plt.show()
```
